```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

A função cos(t) é dita par, pois $cos(t) = cost(-t)$. Portanto, sua componente par é o próprio cosseno e a componente ímpar é zero. Contudo, ao realizar o deslocamento desta função, as suas componentes tem um comportamento diferente da função original. A figura a seguir ilustra que a função $cos(t)$ é par.

```python
_, axes = plt.subplots()
t = np.arange(-4*np.pi,4*np.pi, 0.1)

axes.axvline(x = 0, color = 'red', linestyle = '--')
axes.plot(t,np.cos(t))
```

Ao realizar o deslocamento do sinal de $t+1$, ou seja, considerar o novo sinal $cos(t+1)$. Este, deixa de ser par para o eixo de referência da figura anterior.

```python
_, axes = plt.subplots()
axes.plot(t, np.cos(t+1))
axes.axvline(x = 0, color = 'red', linestyle = '--')
```

Todo sinal pode ser decomposto em suas componentes par e ímpar. Tais componentes são dadas por

$$\begin{equation*}x_p(t) = 1/2[x(t) + x(-t)]\end{equation*}$$
$$\begin{equation*}x_i(t) = 1/2[x(t) - x(-t)]\end{equation*}$$

Analisando graficamente as componentes de cos(t+1),

```python
_, axes = plt.subplots()
xp = 1/2*(np.cos(t+1) + np.cos(-t+1))
xi = 1/2*(np.cos(t+1) - np.cos(-t+1))

axes.set_title('Componentes Par e Impar de $cos(t+1)$')
axes.plot(t, np.cos(t+1), label = '$x(t)$')
axes.plot(t, xp, label = '$x_p(t)$')
axes.plot(t, xi, label = '$x_i(t)$')
axes.set_xlabel('t')
axes.set_ylabel('x(t), $x_p(t)$, $x_i(t)$')
axes.legend()
```
